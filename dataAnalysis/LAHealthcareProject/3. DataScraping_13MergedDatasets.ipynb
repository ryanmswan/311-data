

### Scraping/Cleaning/Merging Datasets- Total of 13 Merged Datasets

#### Objectives: 

1) To compare LA vs State of California on quality and quantity of Healthcare

2) To analyse Healthcare Projects, Expenditures, Facilities, Medically Underserved Areas+Populations, Healthcare Shorgates across counties (causation-correlation)

3) Further analysis on how Medically Underserved Areas+Populations, Healthcare Shortages impact the overall well-being of LA community (causation-correlation)

4) To perform geospatial analysis using Open Location Code (OLC) for granular location classification.


#### Details for Total of 13 Merged Datasets- 3 parts

1) Since most statewide datasets have County/City level for location, we will combine with LA datasets to obtain Latitude-Longitude, Zipcodes. 

2) Scraping and Merging will provide us benefits of flexible multi-cross-correlation analysis.

3) Some Columns are renamed for readability and simplicity.

4) Unwanted Features are dropped
